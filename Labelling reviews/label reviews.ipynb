```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download("vader_lexicon")
import pandas as pd
data = pd.read_csv("D:/Semester 8/NLP/IA/sentimentanalyis_1-main/Preprocessing/Pre_Processed_Reviews.csv")
data = data.dropna()
print(data.head())
```

```python
sentiments = SentimentIntensityAnalyzer()
data["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data["new_reviews"]]
data["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data["new_reviews"]]
data["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data["new_reviews"]]
data['Compound'] = [sentiments.polarity_scores(i)["compound"] for i in data["new_reviews"]]
data.head()
```

```python
score = data["Compound"].values
sentiment = []
for i in score:
    if i >= 0.05 :
        sentiment.append('0')
    elif i < 0 :
        sentiment.append('1')
    else:
        sentiment.append('0')
data["Sentiment"] = sentiment
data.head(98)
```

```python
print(data["Sentiment"].value_counts())
```

```python
data.to_csv('labelled_reviews.csv')
```

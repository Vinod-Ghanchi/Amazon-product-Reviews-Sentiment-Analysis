```python
import re
import pandas as pd
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS
import spacy
```

```python
df=pd.read_csv('D:/Semester 8/NLP/IA/sentimentanalyis_1-main/Scrapping Reviews/Scraping reviews.csv')
df.head()
```

```python
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
```

```python
df['reviews']=df['reviews'].fillna("") #eplaced null value with a value like empty string "" or another default value
```

The first pre-processing step weâ€™ll do is transform all reviews in reviews columns into lower case and create a new column new_reviews.

```python
df['new_reviews'] = df['reviews'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df['new_reviews'].head()
```

Next, we will remove punctuation:

```python
df['new_reviews'] = df['new_reviews'].str.replace('[^\w\s]','')
df['new_reviews'].head()
```

Removing emojis and pictographs(media files) in reviews

```python
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags 
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
df['new_reviews'] = df['new_reviews'].apply(lambda x: remove_emoji(x))
```

Removing Stop words like 'a' 'and 'etc

```python
stop = stopwords.words('english')
df['new_reviews'] = df['new_reviews'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df.head(20)
```

LEmmatization (Transforming each word into its root word)

```python
def space(comment):
    doc = nlp(comment)
    return " ".join([token.lemma_ for token in doc])
df['new_reviews']= df['new_reviews'].apply(space)
df.head(20)
```

```python
df.to_csv("Pre_Processed_Reviews.csv") #saving into csv
```
